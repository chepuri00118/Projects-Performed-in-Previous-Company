```python
dummy=pd.get_dummies(df1['ArtworkCategory'])
dummy.head(5)
```

```python
df2=pd.concat([df1,dummy],axis=1)
df2.head()
```

```python
x=df2.iloc[:,[13,14,15,17,18,19,20,21,22]].values
y=df2.iloc[:,[16]].values
```

```python
#splitting data set into training and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_test
```

# Logistic Regression 

```python
from sklearn.linear_model.logistic import LogisticRegression
lgsr=LogisticRegression(random_state=0)
lgsr.fit(x_train,y_train)
```

```python
y_pred=lgsr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix=confusion_matrix(y_pred,y_test)
conf_matrix
```

```python
from sklearn.metrics import accuracy_score
acc_sco=accuracy_score(y_test,y_pred)
acc_sco
```

# SUPPORT VECTOR CLASSIFICATION(SVC)

```python
#Support vector classification
from sklearn.svm import SVC
sc =SVC(kernel='rbf')
sc_classifier=sc.fit(x_train,y_train)
```

```python
#Predicting Test and Train data
svc_y_test=sc_classifier.predict(x_test)
print(svc_y_test)
```

```python
#Build a confusion matrix on test and tarin
from sklearn.metrics import confusion_matrix
svc_cm_test=confusion_matrix(y_test,svc_y_test)
svc_cm_test
```

```python
#Accuracy classification score
from sklearn.metrics import accuracy_score
svc_acc_test=accuracy_score(y_test,svc_y_test) 
svc_acc_test
```
